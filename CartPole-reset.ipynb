```python
import gym
```

```python
# モデルの読み込み
env = gym.make("CartPole-v0")
```

```python
# 棒が倒れないようにする
# aはランダム
for i_episode in range(20):
    observation = env.reset()
    for t in range(100):
        env.render()
        print(observation)

        # 環境に提案
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)

        if done:
            print("このエピソードは，{}回目で終了しました．".format(t + 1))
            break
```
